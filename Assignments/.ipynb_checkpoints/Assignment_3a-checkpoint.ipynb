# Background

The idea of this assignment is to give you an experience closer to that which you may encounter outside of a teaching setting. You'll work with a dataset that has undergone a lot less processing, label files that are not as clean, and you can expect your results to be worse than a 50% accuracy. Though of course you might get better results. 

That being said, both preprocessing of the images and a fair amount of work on the label files has been done. It is also unlikely that you will get this type of dataset with almost no time to make it work.

This dataset was generated in the Pharmaceutical Bioinformatics group at Uppsala University*. It consists of U-2 OS cells that have been imaged on a single cell level using bright field microscopy. The purpose of the project was to test the laboratory techniques, see if the single cell level images were a good level to analyse the effect of compounds on, as well as do a modelling of how the process might look when applied to blood cells. 

The bright field (bf) images are saved in a 16-bit format compared to the 8-bit format we had in the labs. This means that instead of normalising to 255 we need to normalise to 65535. (2^8 = 256, 2^16 = 65536, 0 is counted as black)

\* This is the group giving this course, and that I'm apart of.

### Code to merge images and masks, you'll need to update it slightly

```python
# From XinYi's masters project
# multiply the bf images with mask to get rid of the backgtound

image_files = [f for f in sorted(os.listdir(images_path)) if f.endswith('.tif')]
imgs = [io.imread(images_path+ str(f)) for f in image_files]

mask_files = [f for f in sorted(os.listdir(masks_path)) if f.endswith('.tif')]
masks = [io.imread(masks_path + str(f)) for f in mask_files]
```

```python
# From XinYi's masters project, slightly modified by Ebba
output_path = ? 
only_black_masks = []

if not os.path.exists(output_path):
    os.mkdir(output_path)
        
for i in range(len(image_files)):
    img_name = image_files[i]
    filename = str(img_name).split('.')
    imgnp = np.array(imgs[i]) #imgnp = imgnp/imgnp.max()
    masknp= np.array(masks[i]) # masknp= masknp/masknp.max()
    
    if imgnp.shape != masknp.shape:
        print("could not merge following file due to shape issues: " + str(filename))
        continue
    if masknp.max() == 0:
        only_black_masks.append(filename)
        continue
   
    masked_np = imgnp*masknp
    masked_np = masked_np/masked_np.max()
    masked_np = img_as_uint(masked_np)
    outfile = output_path + filename[0] + '.tif'
    io.imsave(outfile, masked_np)
    
prin("The following files had masks without any segmentation for the cell, and they are not included in the masked cells: " + str(only_black_masks))
```
